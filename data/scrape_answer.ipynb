```python
import csv
import g4f
import time
import pandas as pd
import os
```

```python

df =  pd.read_csv("business_questions.csv")
df
```

```python
pre = "Give me proffessional answer to "
post = " print answer only no extra text"
```

```python
questions = []
answer = []
count = 0

for i in df["Question"][981:]:
    question = pre + i + post
    
    from g4f.Provider import (  
        ChatgptAi
    )

    # Set with provider
    response = g4f.ChatCompletion.create(
        model="gpt-3.5-turbo",
        provider=g4f.Provider.You,
        auth='cookie or token or jwt ...',
        messages=[{"role": "user", "content": question}],
        # stream=True,
    )

    
    
    string = "".join(map(str,response))
    count+=1
    print(count)

    with open("qa_dataset.csv", 'a', newline='') as csv_file:
        csv_writer = csv.writer(csv_file)
        csv_writer.writerow([i, string])



```
